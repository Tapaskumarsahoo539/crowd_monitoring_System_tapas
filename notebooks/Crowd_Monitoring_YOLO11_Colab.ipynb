### Step 6: Inference and Crowd Monitoring
Test the trained model (or the base `yolo11n.pt`) on a sample image or video.

```python
import cv2
import matplotlib.pyplot as plt

# Load the best weights
# model = YOLO('runs/train/crowd_model/weights/best.pt')
model = YOLO('yolo11n.pt') # Using base model for demonstration

# Perform inference on an image (download a sample image first)
!wget -O sample.jpg "https://images.unsplash.com/photo-1506157786151-b8491531f063?ixlib=rb-4.0.3&auto=format&fit=crop&w=800&q=80" 

results = model.predict('sample.jpg', classes=[0], conf=0.3)

person_count = len(results[0].boxes)
overcrowded = person_count > 20

annotated_frame = results[0].plot()

plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
plt.title(f'Crowd Count: {person_count} | Overcrowded: {overcrowded}')
plt.axis('off')
plt.show()
```
